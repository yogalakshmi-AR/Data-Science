This project attempts to visualize the Netflix stock data. In finance, a stock profile is a series of studies, visualizations, and analyses that dive into different aspects of a publicly traded company's data.

Note - Dow Jones Industrial Average is used to compare the Netflix stock to the larger stock market.

In Yahoo, Adj Close is documented as adjusted close price adjusted for both dividends and splits.
This means this is the column with the true closing price, so these data are very important.

```python
netflix_stocks.rename(columns = {'Adj Close' : 'Price'}, inplace = True)
dowjones_stocks.rename(columns = {'Adj Close' : 'Price'}, inplace = True)
netflix_stocks_quarterly.rename(columns = {'Adj Close' : 'Price'}, inplace = True)
```

Visualising distribution of Netflix stock prices for every quarter in 2017

```python
ax = sns.violinplot(data = netflix_stocks_quarterly, x = 'Quarter', y = 'Price')
ax.set_title("Distribution of 2017 Netflix Stock Prices by Quarter")
ax.set_ylabel("Closing Stock Price")
ax.set_xlabel("Business Quarters in 2017")
plt.show()
```

The plot shows high distribution within the range of 150 and 100.
Q3 has the longest range compared to other quarters.

Visualising estimated vs actual stock prices

```python
x_positions = [1, 2, 3, 4]
chart_labels = ["1Q2017","2Q2017","3Q2017","4Q2017"]
earnings_actual =[.4, .15,.29,.41]
earnings_estimate = [.37,.15,.32,.41 ]
plt.scatter(x_positions, earnings_actual, color='red', alpha=0.5)
plt.scatter(x_positions, earnings_estimate, color='blue', alpha=0.5)
plt.legend(["Actual", "Estimate"])
plt.xticks(x_positions, chart_labels)
plt.title("Earnings Per Share in Cents")
plt.show()
```

Estimated stock prices matched the actual stock prices for two quarters while the actual price exceeds the estimate during Q1 and falls below the estimate in the Q3.

Visualising company's revenue and earnings in 2017

```python
# The metrics below are in billions of dollars
revenue_by_quarter = [2.79, 2.98,3.29,3.7]
earnings_by_quarter = [.0656,.12959,.18552,.29012]
quarter_labels = ["2Q2017","3Q2017","4Q2017", "1Q2018"]

# Revenue
n = 1 # This is our first dataset (out of 2)
t = 1 # Number of dataset
d = 4 # Number of sets of bars
w = 0.5 #dth of each bar
bars1_x = [t*element + w*n for element
             in range(d)]


# Earnings
n = 1 # dataset (out of 2)
t = 1 #Number of dataset
d = 4 #Number of sets of bars
w = 0.5 #th of each bar
bars2_x = [t*element + w*n for element
             in range(d)]






middle_x = [ (a + b) / 2.0 for a, b in zip(bars1_x, bars2_x)]
labels = ["Revenue", "Earnings"]
plt.bar(bars1_x, revenue_by_quarter)
plt.bar(bars2_x, earnings_by_quarter)
plt.title("Revenue and Earnings by Quarter")
plt.xticks(middle_x, quarter_labels)
```

Revenue and earnings follow an upward curve thereby gradually increasing in each quarter. 
Now let's compare Netflix's stock performance with Dow Jones Industrial average to have better insights.

```python
# Left plot Netflix
# ax1 = plt.subplot(total rows, total columns, subplot to modify)



ax1 = plt.subplot(1, 2, 1)
ax1.set_xlabel('Date')
ax1.set_ylabel('Stock Price')
ax1.set_title("Netflix")
plt.xticks(rotation='vertical')

plt.plot(netflix_stocks['Date'], netflix_stocks['Price'], color="purple")

# Right plot Dow Jones
# ax2 = plt.subplot(total rows, total columns, subplot to modify)

ax2 = plt.subplot(1, 2, 2)
ax2.set_xlabel('Date')
ax2.set_ylabel('Stock Price')
ax2.set_title("Dow Jones")

plt.xticks(rotation='vertical')
plt.plot(dowjones_stocks['Date'], dowjones_stocks['Price'], color="green")
plt.subplots_adjust(wspace=.5)
plt.show()
```

Though the stock prices are voltatile, Netflix performs better than Dow Jones Industrial Average in 2017.
